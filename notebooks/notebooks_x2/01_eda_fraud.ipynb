```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configuración para visualizaciones
plt.style.use('seaborn')
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)

# Cargar los datos usando la ruta absoluta
DATA_PATH = r'D:\Proyectos Personales ML\Fraud detection\data\raw\creditcard.csv'
df = pd.read_csv(DATA_PATH)

# Análisis inicial básico
print("Dimensiones del dataset:", df.shape)
print("\nPrimeras 5 filas del dataset:")
display(df.head())

# Información del dataset
print("\nInformación del dataset:")
display(df.info())

# Estadísticas básicas
print("\nEstadísticas descriptivas:")
display(df.describe())

# Distribución de clases (fraudulentas vs no fraudulentas)
print("\nDistribución de clases (0: Normal, 1: Fraude):")
class_distribution = df['Class'].value_counts(normalize=True)
print(class_distribution)

# Visualización del desbalance
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Class')
plt.title('Distribución de Transacciones Normales vs Fraudulentas')
plt.show()
```

```python
#Análisis del desbalance
print("\nDistribución de clases:")
print(df['Class'].value_counts(normalize=True) * 100)

#Estadísticas descriptivas para Amount y Time
print("\nEstadísticas de Amount y Time:")
display(df[['Time', 'Amount']].describe())
```

```python
#Visualización de la distribución de Amount
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Amount'], bins=50)
plt.title('Distribución de Amount')
plt.xlabel('Amount')
plt.yscale('log')

plt.subplot(1, 2, 2)
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Amount por Clase')
plt.show()
```

```python
# Matriz de correlación
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Matriz de Correlaciones')
plt.show()

# Top 10 correlaciones con la clase fraudulenta
correlations = correlation_matrix['Class'].sort_values(ascending=False)
print("\nTop 10 correlaciones con Fraude:")
display(correlations[1:11])  # Excluimos la autocorrelación
```

```python
# Convertir Time a horas para mejor interpretación
df['Hour'] = df['Time'] / 3600

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Hour', bins=24)
plt.title('Distribución de Transacciones por Hora')

plt.subplot(1, 2, 2)
sns.histplot(data=df[df['Class']==1], x='Hour', bins=24, color='red')
plt.title('Distribución de Fraudes por Hora')
plt.show()
```
